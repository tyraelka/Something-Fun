### Video download

```python
from pytube import YouTube

url = "https://www.youtube.com/watch?v=7QZXyHlMhx4"
file = YouTube(url).streams.first().download()
```

* Check the file name

```python
print(file)
```

### Convert video to mp3

```python
from moviepy.editor import VideoFileClip
video = VideoFileClip(file)
music_name = file.split("\\")[-1].replace(".mp4", ".mp3")
video.audio.write_audiofile(music_name)
```
